# 2.
Escriba una función que se llame `A_sin()` que reciba como parámetro a `t` y argumentos `A = 1`, `omega = 2*np.pi` y `phi = np.pi/2` y que regrese $f(t) = A\sin(t\omega_0 + \phi )$. Nota: La función debe contar con un *docstring*. En las celdas a continucación se muestra el funcionamiento de `A_sin()`

```python
def A_sin(t, A = 1, omega = 2*np.pi, phi = np.pi/2): #definimos la función con 1 parámetro y 2 argumentos
    
    """ Regresa la posición de una partícula que oscila en función sinusoidal del tiempo """
    
    return A*(np.sin(t*omega + phi)) 
```

Una función que cuenta con infinitas derivadas y además estás son continuas, se puede escribir como la siguiente serie alrededor de un punto $x_0$ del dominio de la función:

$$f(x) = \displaystyle \sum_{k = 0}^{\infty}\dfrac{f^{k}(x_0)}{k!}(x - x_0)^k$$

Donde $f^{n}(x_0)$ denota la n-ésima derivada de la función $f$ evaluada en $x_0$. Sin embargo podemos aproximar a la función $f$ truncando la serie a un cierto orden $n$ y lo que nos queda es un polinomio de orden $n$, a continuación se muestran algunas aproximaciones de ciertas funciones:

* $\exp(x)  \approx \displaystyle \sum_{k = 0}^{n}\dfrac{x^k}{k!}$

* $\ln(1- x)  \approx - \displaystyle \sum_{k = 1}^{n}\dfrac{x^k}{k}$

* $\sin(x)  \approx \displaystyle \sum_{k = 0}^{n}\dfrac{(-1)^k}{(2k +1)!}x^{2k+1}$


* $\cos(x)  \approx \displaystyle \sum_{k = 0}^{n}\dfrac{(-1)^k}{(2k )!}x^{2k}$

Implemente cuatro funciones cuyos nombres sean: `my_exp()`, `my_ln()`, `my_sin()` y `my_cos()` que aproximen las respectivas funciones a un orden $n$ alrededor de un punto $x_0$. Por ejemplo:

```python
def my_factorial(n): #definimos la función
    
    """ Regresa el factorial de cualquier número n """
    
    factorial_value = 1 #le asignamos a la variable factorial_value el valor de 1
    
    for i in range(1, n + 1): #definimos el rango de 1 a n+1
        factorial_value = factorial_value * i #le asignamos un nuevo valor a la variable multiplicandolo por i
        i += 1
    return factorial_value #regresa factorial_value
```

```python
def my_exp(x, n = 1): #definimos la función
    
    """ Regresa la función exponencial de cualquier número x a partir de la suma de su polinomio """
    
    aux = 0 #le asignamos al auxiliar el valor de 0
    
    for i in range (0, n): #definimos el rango de 0 a n
        aux = aux + ((x**i)/my_factorial(i)) #le asignamos un nuevo valor a aux sumandole la operación
        i += 1
    return aux #regresa aux
```

```python
my_exp(1, 7)  #invocamos la función my_exp con los valores 1 y 7
```

```python
def my_ln(x, n = 1): #definimos la función
    
    """ Regresa el logaritmo natural de cualquier número x a partir de la suma de su polinomio """
    
    x = 1 - x #le asignamos un nuevo valor a x restandolo a 1
    aux = 0 #le asignamos al auxiliar el valor de 0
    
    for i in range (1, n): #definimos el rango de 1 a n
        aux = aux - ((x**i)/i) #le asignamos un nuevo valor a aux restandole la operación
        i += 1
    return aux #regresa aux
```

```python
my_ln(2, 53) #invocamos la función my_ln con los valores 2 y 53
```

```python
def my_cos(x, n = 1): #definimos la función
    
    """ Regresa el coseno de cualquier número x a partir de la suma de su polinomio """

    aux = 0 #le asignamos al auxiliar el valor de 0
    
    for i in range (0, n): #definimos el rango de 0 a n
        aux = aux + ((((-1)**i)/my_factorial(2*i))*x**(2*i)) #le asignamos un nuevo valor a aux sumandole la operación
        i += 1
    return aux #regresa aux
```

```python
my_cos(np.pi/2, 3)  #invocamos la función my_cos con los valores pi/2 y 3
```

```python
def my_sin(x, n = 1): #definimos la función
    
    """ Regresa el seno de cualquier número x a partir de la suma de su polinomio """

    aux = 0 #le asignamos al auxiliar el valor de 0
    
    for i in range (0, n): #definimos el rango de 0 a n
        aux = aux + ((((-1)**i)/my_factorial(2*i + 1))*x**(2*i+1)) #le asignamos un nuevo valor a aux sumandole la operación|
        i += 1
    return aux #regresa aux
```

```python
my_sin(np.pi/2, 3) #invocamos la función my_sin con los valores pi/2 y 3
```

En la celda anterior se está evaluando la función seno en $\frac{\pi}{2}$ a orden $3$, y observe que calcula una aproximación al $\sin(\frac{\pi}{2})$ que es $1$.

```python
my_sin(np.pi/2, 9) #invocamos la función my_sin con los valores pi/2 y 9
```

En la celda anterior se está evaluando la función seno en $\frac{\pi}{2}$ a orden $9$. Y está aproximación en mucho mejor que en el caso anterior
